```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def parselines(filename):
    with open(filename ,"r") as f:
        f.readline()
        lines = []
        for l in f.readlines():
            l = l.strip()
            l = l.split()
            lines.append(l)
    return lines
lines = parselines("resnet_d50_c32_res.txt")
```

```python
def draw_flop_acc(lines):
    x = []
    y1 = []
    y5 = []
    data = []
    for l in lines:
        if( not l[1]=="0.5"):
            continue
        data.append([float(l[2]),float(l[3]),float(l[4])])
        
    data.sort()
#         x.append(l[2])
#         y1.append(l[3])
#         y5.append(l[4])
    data = np.array(data)
#     print(data)
    x = data[:,0]
    y1 = data[:,1]
    y5 = data[:,2]
    plt.plot(x,y1)
    plt.plot(x,y5)
    plt.xlabel("flops")
    plt.xlim((0.2e10,0.9e10))
    plt.ylabel("accuracy")
    plt.ylim((65,100))
    plt.show()
draw_flop_acc(lines)
```

```python
def read_and_draw(filename):
    lines = parselines(filename)
    draw_flop_acc(lines)
read_and_draw("resnet_d50_c32_res.txt")
```
